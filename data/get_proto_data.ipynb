```python
import json
import re

def extract_finish_info(row):
    """Extract thought, action, goal_type, and is_owner from messages"""
    thought = ""
    action = ""
    goal_type = ""
    is_owner = None  # Will be True/False/None
    
    conv = row['conversation']
    success = row['success']
    
    # Extract is_owner from the first message (prompt)
    if conv and len(conv) > 0:
        first_message = conv[0]
        if isinstance(first_message, dict) and 'value' in first_message:
            first_content = str(first_message['value'])
            
            # Look for "owners" or "others" in the first message
            if 'owners' in first_content.lower():
                is_owner = True
            elif 'others' in first_content.lower():
                is_owner = False
            # If neither is found, is_owner remains None
    
    # Check if we should look for finish info
    should_extract = False
    
    if success == 1:
        # If success=1, look at the last message
        should_extract = True
        target_message = conv[-1] if conv else None
    else:
        # Otherwise, check if last message contains "Action:finish"
        if conv:
            last_message = conv[-1]
            if isinstance(last_message, dict) and 'value' in last_message:
                if 'Action:finish' in last_message['value']:
                    should_extract = True
                    target_message = last_message
    
    if should_extract and target_message:
        if isinstance(target_message, dict) and 'value' in target_message:
            value_content = str(target_message['value'])
            
            # Extract thought (before "Action:")
            thought_match = re.search(r'Thought:\s*(.*?)(?=\nAction:|Action:|$)', value_content, re.DOTALL)
            if thought_match:
                thought = thought_match.group(1).strip()
            
            # Extract action (from "Action:" up to but not including "goal_type")
            action_match = re.search(r'Action:\s*(.*?)(?="goal_type"|$)', value_content, re.DOTALL)
            if action_match:
                action = action_match.group(1).strip()
                # Clean up any trailing comma and whitespace from JSON structure
                action = re.sub(r',\s*$', '', action)
            
            # Extract goal_type from JSON structure
            try:
                json_match = re.search(r'\{[^}]*"goal_type"[^}]*\}', value_content)
                if json_match:
                    json_str = json_match.group()
                    parsed = json.loads(json_str)
                    if 'goal_type' in parsed:
                        goal_type = parsed['goal_type']
            except:
                # If JSON parsing fails, try regex
                goal_type_match = re.search(r'"goal_type"\s*:\s*"([^"]*)"', value_content)
                if goal_type_match:
                    goal_type = goal_type_match.group(1)
    
    return thought, action, goal_type, is_owner

# Apply the function to create the four new columns
result = qwen.apply(lambda row: pd.Series(extract_finish_info(row)), axis=1)
qwen["thought"] = result[0]
qwen["action"] = result[1] 
qwen["goal_type"] = result[2]
qwen["is_owner"] = result[3]
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

def extract_first_prompt(conversation):
    """Extract the first human message from conversation"""
    if conversation and len(conversation) > 0:
        first_message = conversation[0]
        second_message = conversation[1]
        if isinstance(first_message, dict) and first_message.get('from') == 'human' and isinstance(second_message, dict) and second_message.get('from') == 'human':
            # Return the message in the format expected by RLHFDataset
            return [{"role": "user", "content": first_message.get('value', '')},
                    {"role": "user", "content": second_message.get('value', '')}]
    return []

# Create new dataframe with just the prompts
prompt_df = pd.DataFrame({
    'prompt': qwen["conversation"].apply(extract_first_prompt),
    'item_id': int(qwen['item_id'].replace('gmail_','')),  # Keep item_id for reference
    'success': qwen['success'],  # Keep success for reference
    'reward': qwen['reward'],     # Keep reward for reference
    'is_owner': qwen['is_owner']     # Keep reward for reference
})

# Filter out empty prompts
prompt_df = prompt_df[prompt_df['prompt'].apply(lambda x: len(x) > 0)]

print(f"Original conversations: {len(qwen)}")
print(f"Valid prompts extracted: {len(prompt_df)}")

train_df, val_df = train_test_split(
    prompt_df,
    test_size=0.05,
    random_state=42,
    shuffle=True
)

# Save to parquet
train_df.to_parquet('./_grpo/train.parquet', index=False)
val_df.to_parquet('./_grpo/val.parquet', index=False)

# Preview the result
print("\nSample prompts:")
print(train_df['prompt'].head(3).values)
```

```python
def count_words_in_prompt(prompt_entry):
    # prompt_entry is something like [{'content': "...", 'role': 'user'}, ...]
    if isinstance(prompt_entry, list):
        # Join all 'content' fields together
        full_text = " ".join(str(msg.get("content", "")) for msg in prompt_entry)
        # Split into words by whitespace and count
        return len(full_text.split())
    return 0  # fallback if unexpected type

# Apply function to get word count for each prompt
df["prompt_word_count"] = df["prompt"].apply(count_words_in_prompt)

# Example: show first few
print(df[["prompt", "prompt_word_count"]].head())

# If you want a single number for one prompt (first row):
print("First prompt word count:", df["prompt_word_count"].iloc[0])
```
